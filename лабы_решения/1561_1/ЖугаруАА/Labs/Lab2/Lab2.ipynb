```python
# Plotly

import plotly
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
```

```python
surv_df = df.groupby('Pclass')[['Survived']].sum().join(   # Посчитаем число Survived по Pclass
        df.groupby('Pclass')[['Sex']].count()
)
surv_df.columns = ['Survived', 'Age']

# создадим линию Survived
trace0 = go.Scatter(
        x=surv_df.index,
        y=surv_df.Survived,
        name='Survived'
)

# создадим линию Age
trace1 = go.Scatter(
        x=surv_df.index,
        y=surv_df.Age,
        name='Age'
)

# определим массив данных и задаем title графика в layout
data = [trace0, trace1]
layout = {'title': 'Statistics...'}

# создаем объект Figure и визуализируем его
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
plotly.offline.plot(fig, filename='surv_age_pclass_stats.html', show_link=False)
```

```python
# считаем 
age_df = df.groupby('Age')[['Fare']].sum().join(
        df.groupby('Age')[['Survived']].count()
)
age_df.columns = ['Fare', 'Age']
age_df.sort_values('Fare', ascending=False, inplace=True)

# создаем traces для визуализации
trace0 = go.Bar(
        x=age_df.index,
        y=age_df.Fare,
        name='Fare'
)
trace1 = go.Bar(
        x=age_df.index,
        y=age_df.Age,
        name='Age'
)

data = [trace0, trace1]
layout = {'title': 'Statistic...2', 'xaxis': {'title': 'age'}}

fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
# создаем Box trace для каждого Embarked
data = []
for embarked in df.Embarked.unique():
    data.append(
        go.Box(y=df[df.Embarked==embarked].Age, name=embarked)
    )
# визуализируем данные
iplot(data, show_link=False)
```
