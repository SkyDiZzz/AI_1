# Группировка данных цены и возраста(по количеству, среднему, мин, мах)

```python

columns_to_show = ['Fare', 'Age']

df.groupby(['Pclass'])[columns_to_show].describe(percentiles=[])
```

# группировка данных в зависимости от класса и вывод статистики по ценe и возрасту

```python
#группировка данных в зависимости от класса и вывод статистики по цену и возрасту
columns_to_show = ['Fare', 'Age']

df.groupby(['Pclass'])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

# Распределение количества мужчин и женщин по классам

```python
pd.crosstab(df['Pclass'], df['Sex'])
```

# Среднее значение возраста и цен по классу

```python
df.pivot_table(['Age', 'Fare'],
['Pclass'], aggfunc='mean').head(10)
```

# Добавляем столбец билет со скидкой 5%

```python
sale= df['Fare']- df['Fare']*0.05
df.insert(loc=len(df.columns), column='Sale 5', value=sale)
df.head()
```

# Удаляем ненужные столбцы

```python
#удаляем столбцы
df=df.drop(['Sale of 5 percent', 'Sale of percent', 'Sale 5'], axis=1)

df.head()
```

# посмотрим как PClass связан с Embarked

```python

pd.crosstab(df['Pclass'], df['Embarked'], margins=True)
```

# Распределение Embarked с помощью графика

```python
df['Embarked'].value_counts().plot(kind='bar', label='Embarked')
plt.legend()
plt.title('Распределение Embarked');
```

```python
corr_matrix = df.drop(['Name', 'Ticket', 'Cabin',
                      'Pclass'], axis=1).corr()
```

# КОРРЕЛЯЦИЯ КОЛИЧЕСТВЕННЫХ ПРИЗНАКОВ

```python
sns.heatmap(corr_matrix); #КОРРЕЛЯЦИЯ КОЛИЧЕСТВЕННЫХ ПРИЗНАКОВ
```

# Распределение количественных признаков

```python

features = list(set(df.columns)-set(['Survived', 'Ticket', 'Cabin',
                                    'Embarked']))
df[features].hist(figsize=(20,12));
```

# Распределение количественных признаков с помощью pairplot

```python
sns.pairplot(df[features + ['Embarked']], hue='Embarked');
```

# Количество потраченных денег по классам 


```python
_, axes = plt.subplots(1, 2, sharey=True, figsize=(16,6))

sns.boxplot(x='Pclass', y='Fare', data=df, ax=axes[0]);
sns.violinplot(x='Pclass', y='Fare', data=df, ax=axes[1]);
```

# Распределение выживших и погибших по классам
* Выживших в 1 и 3 классе почти равное количество
* Погибших в 3 классе значительно превышает по количеству 1го и 2го класса

```python
sns.countplot(x='Survived', hue='Pclass', data=df);
```

# Доля выжевших по классам
* Видно что в 1ом классе процент выжевших максимален , в 3ем минимален
